# Time Series Analysis

```python
ashish_david_df['year'] = ashish_david_df['year'] + 1900
ashish_david_df['year'] = pd.to_datetime(ashish_david_df['year'], format='%Y')
ashish_david_df.set_index('year', inplace=True)
```

```python
sns.set(rc={'figure.figsize':(11, 4)})
plt.rcParams['figure.figsize'] = (8,5)
plt.rcParams['figure.dpi'] = 150
```

```python
# Checking the income growth trends over years
ashish_david_df['gincpc'].plot(marker=".",linewidth=0.25) # gincpc represents the growth rate of per capita income
plt.title('Time Series Plot for gincpc (21BDS0335)')
plt.xlabel('Year')
plt.ylabel('gincpc')
plt.show()
```

```python
ashish_david_cols = ashish_david_df.select_dtypes(include=['number']).columns
ashish_david_df_yearly = ashish_david_df[ashish_david_cols].groupby(ashish_david_df.index).mean()
ashish_david_selected_features = ['pris', 'criv', 'crip', 'unem', 'incpc']
```

```python
fig, ashish_david_axes = plt.subplots(nrows=len(ashish_david_selected_features), ncols=1, figsize=(10, 12), sharex=True)
for i, col in enumerate(ashish_david_selected_features):
    ashish_david_axes[i].plot(ashish_david_df_yearly.index, ashish_david_df_yearly[col],marker="*",linestyle='-', label=col)
    ashish_david_axes[i].set_ylabel(col)
plt.suptitle("Time Series Trends (21BDS0335)", fontsize=14)
plt.xlabel("Year")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

features = ['pris','polpc','black','metro','crip','incpc','unem','ag0_14','ag25_34']
target = 'criv'
df = ashish_david_df

# Define X (independent) and y (dependent variable)
X = df[features]
y = df[target]

# Split into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Linear Regression model
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train, y_train)

# Predictions
y_pred = rf_regressor.predict(X_test)

# Model Evaluation
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"RMSE: {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
plt.scatter(y_test, y_pred, alpha=0.6)
plt.xlabel("Actual criv")
plt.ylabel("Predicted criv")
plt.title("Actual vs Predicted Violent Crime Rate (criv)")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.show()
```
