```python
df=pd.read_csv("prison.csv")
df.head(5)
```

```python
print("Dataset Dimensions:", df.shape)
```

```python
print("Summary:")
display(df.describe(include='all'))
```

```python
print("Checking for missing values:")
display(df.isnull().sum())
```

```python
print("Total number of missing values:")
display(df.isnull().sum().sum())
```

```python
print("Data Types:")
display(df.dtypes)
```

```python
# Range (Max - Min)
numeric_cols = df.select_dtypes(include=['number'])
print("Range:")
print(numeric_cols.max() - numeric_cols.min())
```

```python
# Interquartile Range (IQR)
print("Interquartile Range (IQR):")
print(numeric_cols.quantile(0.75) - numeric_cols.quantile(0.25))
```

```python
# Interdecile Range
print("Interdecile Range:")
print(numeric_cols.quantile(0.90) - numeric_cols.quantile(0.10))
```

```python
# Standard Deviation
print("Standard Deviation:")
print(numeric_cols.std())
```

```python
# Skewness
print("Skewness:")
print(numeric_cols.skew())
```

```python
# Kurtosis
print("Kurtosis:")
print(numeric_cols.kurt())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Univariate Analysis
numeric_cols = df.select_dtypes(include=[np.number]).columns[:5]
plt.figure(figsize=(12,6))
df[numeric_cols].hist(figsize=(15, 10), bins=20, layout=(4,3))
plt.suptitle("Distribution of Selected Numeric Features")
plt.show()
```

```python
# Bivariate Analysis
corr_matrix = df[numeric_cols].corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix (Subset of Features)")
plt.show()
```

```python
# Multivariate Analysis
sns.pairplot(df[numeric_cols])
plt.show()
```

```python
#Boxplot
plt.figure(figsize=(12,6))
sns.boxplot(data=df[numeric_cols])
plt.xticks(rotation=90)
plt.title("Boxplot for Outlier Detection (Subset)")
plt.show()
```

```python
# Time Series Data Analysis
df['year'] = df['year'] + 1900
df['year'] = pd.to_datetime(df['year'], format='%Y')
df.set_index('year', inplace=True)
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns
df_yearly = df[numeric_cols].groupby(df.index).mean()
selected_features = ['pris', 'criv', 'crip', 'unem', 'incpc']
fig, axes = plt.subplots(nrows=len(selected_features), ncols=1, figsize=(10, 12), sharex=True)
for i, col in enumerate(selected_features):
    axes[i].plot(df_yearly.index, df_yearly[col], linestyle='-', label=col)
    axes[i].set_ylabel(col)
    axes[i].legend()
    axes[i].grid()
plt.suptitle("Time Series Trends (21BDS0335)", fontsize=14)
plt.xlabel("Year")
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
